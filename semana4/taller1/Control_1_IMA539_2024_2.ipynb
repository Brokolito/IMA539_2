```python
# Respuesta Pregunta 1
import pandas as pd
df=pd.read_csv("Base_de_datos_control_1.csv")
df.describe(include="all")
```

```python
# Respuesta Pregunta 2
df.info()
X=df.drop(columns=["Action"]).copy()
y = df.iloc[:,4]
df.head(4)
```

```python
# Respuesta Pregunta 3
clases = y.nunique()
nro_clases = y.value_counts()

print(f"Número de clases distintas: {clases}")

print("Número de datos por cada clase:")

print(nro_clases)
```

```python
# Respuesta Pregunta 4
import matplotlib.pyplot as plt
import numpy as np
X.describe(include="all")
X=X.iloc[:,0:2]
X.describe()
y.unique()
y.loc[y == "allow"] = 0
y.loc[y == "drop"] = 1
y.loc[y == "deny"] = 2
y=y.astype(int)
y.describe()
```

```python
# Respuesta Pregunta 5
X_std = np.copy(X)
X_std[:, 0] = (X_std[:, 0] - X_std[:, 0].mean()) / X_std[:, 0].std()
X_std[:, 1] = (X_std[:, 1] - X_std[:, 1].mean()) / X_std[:, 1].std()

```

```python
# Respuesta Pregunta 6
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=37, stratify=y)
print("Dimensiones de X_train:", X_train.shape)
print("Dimensiones de X_test:", X_test.shape)
print("Dimensiones de y_train:", y_train.shape)
print("Dimensiones de y_test:",y_test.shape)
```

```python
# Respuesta Pregunta 7
from sklearn.linear_model import Perceptron
y_per = y_train.copy() 
y_per[(y == 1) | (y == 2)] = 0
y_per[y == 0] = 1
ppn_skl = Perceptron(max_iter=20, alpha=0.1, random_state=1, n_jobs=-1)
ppn_skl.fit(X_train, y_per)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1.0, random_state=1)
lr.fit(X_train, y_train)
```
